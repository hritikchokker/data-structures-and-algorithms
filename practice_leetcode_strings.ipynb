# Remove Duplicate Letters

Given a string s, remove duplicate letters so that every letter appears once and only once. You must make sure your result is the smallest in lexicographical order among all possible results.

 

Example 1:

Input: s = "bcabc"
Output: "abc"

Example 2:

Input: s = "cbacdcbc"
Output: "acdb"


```python
class Solution:
    def removeDuplicateLetters(self,s):
        dic={}
        for i in s:
            if i not in dic:
                dic[i]=i
        return ''.join(sorted(dic.values()))
        
sol = Solution()    
print(sol.removeDuplicateLetters('bcabc'))
```

# Find the Difference

You are given two strings s and t.

String t is generated by random shuffling string s and then add one more letter at a random position.

Return the letter that was added to t.

 

Example 1:

Input: s = "abcd", t = "abcde"
Output: "e"
Explanation: 'e' is the letter that was added.

Example 2:

Input: s = "", t = "y"
Output: "y"

Example 3:

Input: s = "a", t = "aa"
Output: "a"

Example 4:

Input: s = "ae", t = "aea"
Output: "a"


```python
from collections import Counter
class Solution:
    def findTheDifference(self, s,t):
        t=Counter(t)
        s=Counter(s)
        res=""
        print(t.items(),'items',s.keys(),'skey')
        for key,value in t.items():
            if key not in s.keys():
                res+=key
                return res
            if t[key] != s[key]:
                res+=key
                return res
        return res        
sol = Solution()    
print(sol.findTheDifference('abcd','abcde')) # e
# print(sol.findTheDifference('','y'))
# print(sol.findTheDifference('a','aa'))
```

#  Longest Substring Without Repeating Characters

Given a string s, find the length of the longest substring without repeating characters.

 

Example 1:

Input: s = "abcabcbb"
Output: 3
Explanation: The answer is "abc", with the length of 3.

Example 2:

Input: s = "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.

Example 3:

Input: s = "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3.
Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.

Example 4:

Input: s = ""
Output: 0


```python
class Solution:
    def lengthOfLongestSubstring(self, s):
        st=set()
        left=0
        res=0
        n=len(s)
        for r in range(n):
            while s[r]  in st:
                st.remove(s[left])
                left+=1
            st.add(s[r])
            res=max(res,r-left+1)
        return res
sol = Solution()    
print(sol.lengthOfLongestSubstring('abcabcbb')) # 3
print(sol.lengthOfLongestSubstring('pwwkew')) # 3
print(sol.lengthOfLongestSubstring('au')) # 2
```
