# linked list implementation in python3

```python
"""
    Creates a node with data and assign next to None
"""
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class LinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.next=None
    def append(self,data):
        head=Node(data)
        if self.head is None:
            self.head=head
            self.tail=head
            self.next=None
            return
        else:
            self.tail.next=head
            self.tail=self.tail.next
    """ inserts a new item/element in the linked list at a given position """
    def insertAtPos(self,head,index,data):
        newNode=Node(data)
        temp=head
        for i in range(0,index-1):
            head=head.next
        newNode.next=head.next
        head.next=newNode
    """ inserts a new item at the beginning of the linkedList"""
    def insertAtBeginning(self,head,data):
        newNode=Node(data)
        if head is None:
            self.head=newNode
            self.tail=newNode
            return data
        newNode.next=head
        self.head=newNode
    def display(self,head):
        while head is not None:
            print('=>',head.data,end=" ")
            head=head.next
    def displayRec(self,head):
        if head is not None:
            print('=>',head.data,end=" ")
            self.displayRec(head.next)
    def countNodes(self,head):
        count=0
        while head is not None:
            count+=1
            head=head.next
        return count
    def countNodesRec(self,head):
        if head is None:
            return 0
        else:
            return self.countNodesRec(head.next)+1
    def sumOfAllNodes(self,head):
        sum=0
        while head is not None:
            sum+=head.data
            head=head.next
        return sum
    def sumOfAllNodesRec(self,head):
        if head is None:
            return 0
        return self.sumOfAllNodesRec(head.next)+head.data
    def maximumElement(self,head):
        if head is None:
            return -1
        max=-9812812 # assign max with smallest integer as per the given list of item
        while head is not None:
            if head.data>max:
                max=head.data
            head=head.next
        return max
    def maximumElementRec(self,head):
        if head is None:
            return -9812021
        else:
            max=self.maximumElementRec(head.next)
            return max if max>head.data else head.data
    def searchElement(self,head,data):
        if head is None:
            return -1
        while head is not None:
            if head.data==data:
                return data
            head=head.next
        return -1
    def searchElementRec(self,head,data):
        if head is None:
            return -1
        if head.data==data:
            return data
        return self.searchElementRec(head.next,data)
#      using move to head technique to optimise search
    def searchOptimized(self,head,data):
        if head is None:
            return -1
        tail=None
        while head is not None:
            if head.data ==data:
                temp=tail
                tail.next=head.next
                head.next=self.head
                self.head=head
                return data
            tail=head
            head=head.next
        return -1    
                
ll=LinkedList()
# [3,5,7,10,15]
arr=[int(x) for x in input().split()]
for i in range(len(arr)):
    ll.append(arr[i])
ll.displayRec(ll.head) 
print()
print(ll.countNodesRec(ll.head),'nodes in LL')
print(ll.sumOfAllNodesRec(ll.head),'is the sum of all nodes')
print(ll.maximumElementRec(ll.head),'is the maximum in ll')
print()
print('find an element from the list')
item=int(input())
print(ll.searchOptimized(ll.head,item),'item status')
ll.displayRec(ll.head)
print()
print('insert at begining')
newItem=int(input())
ll.insertAtBeginning(ll.head,newItem)
ll.display(ll.head)
print()
itemData=int(input('add item at specific position'))
index=int(input('index at'))
ll.insertAtPos(ll.head,index,itemData)
ll.display(ll.head)
# 3 5 7 10 15        
# 3 5 7 9 10 11 15
```
