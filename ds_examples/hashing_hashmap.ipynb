```python
"""
    Creates a node with data and assign next to None
"""
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class LinkedList:
    def __init__(self):
        self.head=None
        self.tail=None
        self.next=None
    def append(self,data):
        head=Node(data)
        if self.head is None:
            self.head=head
            self.tail=head
            self.next=None
            return
        else:
            self.tail.next=head
            self.tail=head
#             self.tail=self.tail.next
    """ inserts a new item/element in the linked list at a given position """
    def insertAtPos(self,head,index,data):
        newNode=Node(data)
        temp=head
        for i in range(0,index-1):
            head=head.next
        newNode.next=head.next
        head.next=newNode
#         self.tail.next=newNode
    """ inserts a new item at the beginning of the linkedList"""
    def insertAtBeginning(self,head,data):
        newNode=Node(data)
        if head is None:
            self.head=newNode
            self.tail=newNode
            return data
        newNode.next=head
        self.head=newNode
        self.tail=newNode
    def display(self,head):
        while head is not None:
            print('=>',head.data,end=" ")
            head=head.next
    def displayRec(self,head):
        if head is not None:
            print('=>',head.data,end=" ")
            self.displayRec(head.next)
    def countNodes(self,head):
        count=0
        while head is not None:
            count+=1
            head=head.next
        return count
    def countNodesRec(self,head):
        if head is None:
            return 0
        else:
            return self.countNodesRec(head.next)+1
    def sumOfAllNodes(self,head):
        sum=0
        while head is not None:
            sum+=head.data
            head=head.next
        return sum
    def sumOfAllNodesRec(self,head):
        if head is None:
            return 0
        return self.sumOfAllNodesRec(head.next)+head.data
    def maximumElement(self,head):
        if head is None:
            return -1
        max=-9812812 # assign max with smallest integer as per the given list of item
        while head is not None:
            if head.data>max:
                max=head.data
            head=head.next
        return max
    def maximumElementRec(self,head):
        if head is None:
            return -9812021
        else:
            max=self.maximumElementRec(head.next)
            return max if max>head.data else head.data
    def searchElement(self,head,data):
        if head is None:
            return -1
        while head is not None:
            if head.data==data:
                return data
            head=head.next
        return -1
    def searchElementRec(self,head,data):
        if head is None:
            return -1
        if head.data==data:
            return data
        return self.searchElementRec(head.next,data)
#      using move to head technique to optimise search
    def searchOptimized(self,head,data):
        if head is None:
            return -1
        tail=None
        while head is not None:
            if head.data ==data:
                temp=tail
                tail.next=head.next
                head.next=self.head
                self.head=head
                return data
            tail=head
            head=head.next
        return -1 
    
        """ insert a new element in a sorted linked list on the correct place """
    def insertInSortedLL(self,head,data):
        newNode=Node(data)
#         if head is None:
#             self.head=newNode
#             self.tail=newNode
#             return
#             1 2 3 4 6 7 8 9
        tail=None
        while head is not None:
            if data<head.data:
                tail.next=newNode
                newNode.next=head.next
                return
            tail=head
            head=head.next
    def deleteAtFirst(self,head):
        if head is None:
            return -1
        head=head.next
        self.head=head
    def deleteAtIndex(self,head,index):
        if head is None:
            return -1
        tail=None
        count=0
        while count<index-1:
            tail=head
            head=head.next
            count+=1
        tail.next=head.next
        return 
    def checkIfSorted(self,head):
        if head is None:
            return -1
        tail=head
        head=head.next
        while head is not None:
            if tail.data >head.data:
                return 'll is not sorted'
            tail=head
            head=head.next
        return 'll is sorted'
    """ removes duplicate from a sorted list """
    def removeDupInSortll(self,head):
        if head is None:
            return -1
        tail=head
        head=head.next
        while head is not None:
            if tail.data == head.data:
                tail.next=head.next
                head=head.next
            else:
                tail=head
                head=head.next    
            """ both approach works for this function """
#             if tail.data == head.data:
#                 tail.next=head.next
#                 head=head.next
#                 continue
#             tail=head
#             head=head.next
    def reverseLLUsingArr(self,head):
        if head is None:
            return -1
        count=0
        temp=head
        while temp is not None:
            count+=1
            temp=temp.next
        lst=[0]*count # will create a array(list) on count and initialise all the the indices value to 0
        temp=head
        count=0
        while temp is not None:
            lst[count]=temp.data
            count+=1
            temp=temp.next
        count-=1
        temp=head
        while temp is not None:
            temp.data=lst[count]
            count-=1
            temp=temp.next
        return
    """ reverse a linked list using sliding pointer technique """
    def reverseSlidingP(self,head):
        if head is None:
            return -1
        q=None
        r=None
        while head is not None:
            r=q
            q=head
            head=head.next
            q.next=r
        head=q
        self.head=head
    """ reversing a linked list using recursion """
    def reverseLLRec(self,head,q=None):
        if head is not None:
            self.reverseLLRec(head.next,head)
            head.next=q
        else:
            head=q
            self.head=head
    def concatLL(self,head,ll):
        if head is None:
            self.head=ll
            self.tail=ll
            return
        while head.next is not None:
            head=head.next
        head.next=ll
    def mergeLL(self,head,second):
        if head.data<second.data:
            third=last=head
            head=head.next 
            third.next=None
        else:
            third=last=second
            second=second.next 
            third.next=None
        """ handling case when both list are not of same size """
        while head is not None and second is not None:
            if head.data<second.data:
                last.next=head
                last=head
                head=head.next
                last.next=None
            else:
                last.next=second
                last=second
                second=second.next
                last.next=None
        if head is not None:
            last.next=head
        if second is not None:
            last.next=second
        return
    """ finds whether a linked list contains loop or not """
    def detectLoop(self,head):
        if head is None:
            return False
        slow=head
        fast=head
        while fast is not None and slow is not None and fast!=slow:
            slow=head.next
            fast=head.next.next
            head=head.next
        if slow ==fast:
            return True
        else:
            return False
```
